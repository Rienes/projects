# 3. *Metrics*

В игре Plants & Gardens каждый месяц проводятся тематические события, ограниченные по времени. В них игроки могут получить уникальные предметы для сада и персонажей, дополнительные монеты или бонусы. Для получения награды требуется пройти ряд уровней за определенное время. С помощью каких метрик можно оценить результаты последнего прошедшего события?

Предположим, в другом событии мы усложнили механику событий так, что при каждой неудачной попытке выполнения уровня игрок будет откатываться на несколько уровней назад. Изменится ли набор метрик оценки результата? Если да, то как?

### Стандартная механика

**DAU/WAU/MAU** — количество уникальных пользователей, которые зашли в приложение в течение деня/недели/месяца. Смотрим как меняется DAU в моменты проведения тематических событий

**Stickness Ratio  = avg. DAU/MAU** — коэффициент того, насколько часто пользователи, в течении одного периода, "остаются" в продукте

**Users Online / MAX Users Online** — число пользователей онлайн и максимальный онлайн за период. Следим за онлайном в игре, сравниваем периоды без событий с периодами, в которые проводились тематические события

**Average Session Length**— средняя продолжительность сессий. Так узнаем, стали ли пользователи проводить больше времени в приложении в момент ивентов

**Paying Users**— количество уникальных платящих пользователей за день

**Transactions by User**— среднее количество транзакций на одного платящего пользователя в течение выбранного периода

### Усложненная механика

Описанные выше метрики подходят и для усложнённой механики игры. Дополнительно было бы интересно посмотреть на распределение игроков по уровням и количество отвалившихся игроков в зависимости от уровня, на котором они остановились
